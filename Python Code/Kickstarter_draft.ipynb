### Discision Tree & Random Forest to predict the status of the test values

```python
features = ['goal', 'pledged','funded percentage','backers','levels','updates','comments','Place','Success_or_not'] #replace with your features
target = 'status'
machine_df = comic3_df[features + [target]].dropna()
numerical_features = ['goal', 'pledged','funded percentage','backers','levels','updates','comments'] #replace with your numerical features
categorical_features = ['Place','Success_or_not'] #replace with your categorical features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import classification_report, confusion_matrix

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(), categorical_features)])
```

```python
# 4. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    machine_df[features], machine_df[target], test_size=0.2, random_state=42)

X_train_transformed = preprocessor.fit_transform(X_train)
X_test_transformed = preprocessor.transform(X_test)
```

```python
# 5. Model Training (Decision Tree)
model = DecisionTreeClassifier(random_state=42)
model.fit(X_train_transformed, y_train)
```

```python
# 6. Model Evaluation
y_pred = model.predict(X_test_transformed)
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred))
print("Decision Tree Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# 6. Model Evaluation
y_pred = model.predict(X_test_transformed)
cm = confusion_matrix(y_test, y_pred)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=model.classes_, yticklabels=model.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
# 2. Feature Selection and Data Cleaning
lst_fea = ['goal', 'backers', 'levels', 'updates', 'comments', 'Place', 'Success_or_not', 'funded percentage']  # Replace with relevant features
str_tar = 'pledged'

#Handle nulls.
df_dat = comic3_df[lst_fea + [str_tar]].dropna()

#Convert Place and Success_or_not to numerical.
df_dat['Place'] = df_dat['Place'].astype(int)
df_dat['Success_or_not'] = df_dat['Success_or_not'].astype(int)

# 3. Preprocessing (Scaling)
lst_num = ['goal', 'backers', 'levels', 'updates', 'comments', 'funded percentage'] #all numerical features.

scl_std = StandardScaler()
df_dat[lst_num] = scl_std.fit_transform(df_dat[lst_num])

# 4. Train-Test Split
X_trn, X_tst, y_trn, y_tst = train_test_split(
    df_dat[lst_fea], df_dat[str_tar], test_size=0.2, random_state=42)

# 5. Model Training (Random Forest Regression)
mod_rf = RandomForestRegressor(random_state=42)
mod_rf.fit(X_trn, y_trn)

# 6. Model Evaluation
y_prd = mod_rf.predict(X_tst)
mse_val = mean_squared_error(y_tst, y_prd)
rmse_val = mean_squared_error(y_tst, y_prd, squared=False)
mae_val = mean_absolute_error(y_tst, y_prd)
r2_val = r2_score(y_tst, y_prd)

print(f"Mean Squared Error (MSE): {mse_val}")
print(f"Root Mean Squared Error (RMSE): {rmse_val}")
print(f"Mean Absolute Error (MAE): {mae_val}")
print(f"R-squared (R2): {r2_val}")

# 7. (Optional) Hyperparameter Tuning (GridSearchCV)
# from sklearn.model_selection import GridSearchCV
# param_grid = {'n_estimators': [100, 200], 'max_depth': [10, 20]}
# grd_src = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=3)
# grd_src.fit(X_trn, y_trn)
# best_mod = grd_src.best_estimator_
# y_prd = best_mod.predict(X_tst)
# print(f"R-squared (R2) after GridSearch: {r2_score(y_tst, y_prd)}")

#Create a table to show the actual and predicted values.
df_res = pd.DataFrame({'Actual Pledged': y_tst, 'Predicted Pledged': y_prd})
print(df_res)
```
